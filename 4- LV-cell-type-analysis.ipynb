```r
files <- list.files(path = "data/input/LV_data",pattern = "*.pkl") # Adjust the pattern to match your file types
ldf <- list()
for (k in 1:length(files)) {
    ldf[[k]] <- pd$read_pickle(paste("data/input/LV_data/",files[k],sep = ""))
    ldf[[k]]$LV <- colnames(ldf[[k]])[2]
    colnames(ldf[[k]])[2] <- "LV_weight"
}

```

```r
LV_cell_T1D <- bind_rows(ldf)
```

```r
str(LV_cell_T1D)
```

```r
LV_cell_T1D %>% 
  filter(LV_weight == max(LV_weight), .by = LV)
```

```r
LV_cell_T1D %>% 
  group_by(`cell type`)%>%
summarise(sum_weight = sum(abs(LV_weight)))%>%
arrange(desc(sum_weight))
```
